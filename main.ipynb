## Second filtering of columns
In questa selezione, le colonne selezionate saranno quelle che presentano la minor correlazione possibile tra di essi e viene utilizzato come criterio il VIF (Variance Inflation Factor)

```python
# Variance Inflation Factor
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(df):
    """
    Calculate VIF
    """
    vif = pd.DataFrame() # Inizializzo un nuovo dataframe
    vif["Variable"] = df.columns # Nella colonna Variable inserisco il nome delle colonne di df
    vif["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])] # Nella colonna VIF calcolo il VIF corrispondente a ogni colonna
    vif = vif.sort_values(by="VIF", ascending=False).reset_index(drop=True) # Ordino il dataframe in ordine decrescente in base a VIF
    return vif

vif = calculate_vif(df) # Calcolo il VIF sul dataset della bancarotta

# Conto le features con un VIF alto
vif_threshold = 5 # Imposto un limite per il VIF
features_to_remove = vif[vif['VIF'] > vif_threshold]['Variable'] # Seleziono le colonne che hanno un VIF maggiore del limite impostato (vif_threshold) e le salvo in una variabile

# Rimuovo le features con VIF > 5
X_uncorr = df.drop(columns = features_to_remove) # Salvo il nuovo dataset nella variabile X_uncorr

vif = calculate_vif(X_uncorr) # Calcolo il VIF sulle colonne rimanenti
print(vif)
print("Features removed:", features_to_remove.shape[0]) # Visualizzo quante colonne sono state rimosse
print("Remaining features:", X_uncorr.shape[1]) # Visualizzo quante colonne sono rimaste
 
```

## EDA - Highlight Outliers
Per individuare gli outliers, verrà utilizzata la tecnica del range interquartile (quella utilizzata dai boxplots). Gli outliers verranno divisi per classi del Target (0,1) e salvati poi in un unico dizionario .

```python
outliers_number = {} # Questo dizionario conterrà il numero di outliers per features
outliers_val = {} # Questo dizionario conterrà il valore degli outliers per feeatures

for col in X_uncorr.columns[:-1]:
     # Quartili per target = 0
    Q1_0 = X_uncorr[X_uncorr['Bancarotta'] == 0][col].quantile(0.25)
    Q3_0 = X_uncorr[X_uncorr['Bancarotta'] == 0][col].quantile(0.75)

    # Intervallo Interquartile per target = 0
    IQR_0 = Q3_0 - Q1_0

    # Limite inferiore e superiore per target = 0
    lower_bound_0 = Q1_0 - 1.5 * IQR_0 
    upper_bound_0 = Q3_0 + 1.5 * IQR_0

    # Quartili per target = 1
    Q1_1 = X_uncorr[X_uncorr['Bancarotta'] == 1][col].quantile(0.25)
    Q3_1 = X_uncorr[X_uncorr['Bancarotta'] == 1][col].quantile(0.75)

    # Intervallo Interquartile per target = 1
    IQR_1 = Q3_1 - Q1_1

    # Limite inferiore e superiore per target = 1
    lower_bound_1 = Q1_1 - 1.5 * IQR_1
    upper_bound_1 = Q3_1 + 1.5 * IQR_1

    # Individuo gli outliers per Target = 0 e li salvo nel dizionario
    outliers_0 = [val for val in X_uncorr[X_uncorr['Bancarotta'] == 0][col] if val < lower_bound_0 or val > upper_bound_0]
    outliers_number[f'{col}_0'] = len(outliers_0) # Salvo il numero di outliers per colonna, con target = 0
    outliers_val[f'{col}_0'] = outliers_0 # Salvo il valore degli outliers per colonna, con target = 0


    # Individuo gli outliers per Target = 1 e li salvo nel dizionario
    outliers_1 = [val for val in X_uncorr[X_uncorr['Bancarotta'] == 1][col] if val < lower_bound_1 or val > upper_bound_1]
    outliers_number[f'{col}_1'] = len(outliers_1) # Salvo il numero di outliers per colonna, con target = 1
    outliers_val[f'{col}_1'] = outliers_1 # Salvo il valore degli outliers per colonna, con target = 1

print("Numero di Outliers per Features")
for val in outliers_number:
    print(f'{val}: {outliers_number.get(val)}')
```

Visualizzo graficamente il numero di outliers per features, divisi per classe di Target.

```python
# Creo prima i dataframe funzionali alla visualizzazione grafica
outliers_0_table = pd.DataFrame(columns = ['Features', 'Outliers']) # Inizializzo la tabella che contiene gli outliers per taget = 0
outliers_1_table = pd.DataFrame(columns = ['Features', 'Outliers']) # Inizializzo la tabella che contiene gli outliers per taget = 1

index_0 = 0 # Indici outliers per target = 0
index_1 = 0 # Indici outliers per target = 1

for key in outliers_number.keys(): # Itero sul dizionario costruito precedentemente
    feature = key[:-2] # Salvo nome della colonna escludendo gli ultimi 2 caratteri (_0/_1)
    value = outliers_number.get(key) # Salvo il numero di outliers per la colonna iterata
    if(key.endswith('0')): # Filtro in base alla classe del target e riempio di valori i rispettivi dataset
        outliers_0_table.loc[index_0] = [feature, value] 
        index_0 += 1
    else:
        outliers_1_table.loc[index_1] = [feature, value]
        index_1 += 1
```

```python
# Visualizzazione grafica
fig = plt.figure(figsize=(12,8))

gs = mpl.gridspec.GridSpec(1, 2) # Crea una griglia con 1 riga e 2 colonne per la disposizione dei subplot

ax1 = plt.subplot(gs[0,0]) # Creo il primo subplot che occupa prima riga e prima colonna
ax2 = plt.subplot(gs[0,1]) # Creo il secondo subplot che occupa prima riga e seconda colonna

# Barplot
sns.barplot(data = outliers_0_table, y = 'Features', x = 'Outliers', ax = ax1, color = 'darkred') # Barplot degli outliers per target = 0
sns.barplot(data = outliers_1_table, y = 'Features', x = 'Outliers', ax = ax2, color = 'darkblue') # Barplot degli outliers per target = 1

# Stile generale
fig.suptitle('Numero di Outliers per Features divisi per classe di Target', fontsize = 16) # Titolo generale
fig.supxlabel('Number of Outliers', fontsize = 14) # Label generale per l'asse x

# Stile subplot di sinistra
ax1.set_title('Target = 0')
ax1.set_xlabel('')
ax1.set_ylabel('Features', fontsize = 14, labelpad = 25)

# Stile subplot di destra
ax2.set_title('Target = 1')
ax2.set_xlabel('')
ax2.set_ylabel('')
ax2.set_yticklabels('')

#plt.savefig('plot/barplot/numero_di_outliers_per_features_divisi_per_classe_di_target_barplot.png')
plt.show()
```

Questa volta, per individuare gli outliers sarà utilizzata la tecnica della deviazione standard. In base a questa tecnica, qualsiasi valore distante tre deviazioni standard dalla media della sua distribuzione rappresenta un outlier. Per individuare gli outliers estremi sarà aumentato il coefficiente di distanza a 15 (valore testato manualmente).

```python
X_uncorr_clean = X_uncorr.copy() # inizializzo un nuovo dataset identico a quello originale

k = 15 # Coefficiente di distanza 
outliers_number = 0 # Inizializzo il numero di outliers

for col in columns_with_outliers: # Itero sulle features con outliers estremi

    # Media e deviazione standard
    mean = X_uncorr[col].mean() 
    std_dev = X_uncorr[col].std()

    # Limite inferiore e superiore
    lower_bound = mean - k * std_dev 
    upper_bound = mean + k * std_dev

    # Salvo gli outliers
    extreme_outliers = [val for val in X_uncorr[col] if val > upper_bound or val < lower_bound]

    # Tengo il conto del numero degli outliers rilevati
    outliers_number += len(extreme_outliers) 

    # Pulisco il dataset eliminando gli outliers rilevati
    for i in range(len(extreme_outliers)):
        X_uncorr_clean = X_uncorr_clean[~(X_uncorr_clean[col] == extreme_outliers[i])]

print(f'Outliers estremi individuati: {outliers_number}')
print(f'Osservazioni rimosse: {X_uncorr.shape[0] - X_uncorr_clean.shape[0]}')
```

Effettuiamo dunque un operazione di undersampling, riducendo la dimensione della classe maggioritaria (0) e renderla uguale a quella minoritaria. In tal modo, assicuriamo una eguale rappresentazione di entrambi le classi 

```python
# Distribuzione delle classi del taget nel dataset pulito dagli outliers
X_uncorr_norm["Bancarotta"].value_counts()
```

```python
y = X_uncorr_norm["Bancarotta"] # Isoliamo la variabile Y

# Estraiamo l'indice delle righe di ciascuna classe
index_class_0 = y[y == 0].index
index_class_1 = y[y == 1].index

np.random.seed(42)  # Impostiamo il seme per la generazione casuale

# Selezioniamo lo stesso numero di righe per entrambi le classi
same_size_index_class_0 = np.random.choice(index_class_0, y.value_counts().min(), replace=False) # Vengono selezionati 214 valori casuali (y.value_counts().min()) tra gli indici della classe 0 (index_class_0 

all_index = np.concatenate([index_class_1, same_size_index_class_0]) # Unisco le due liste di indici
X_uncorr_balanced = X_uncorr_norm.loc[all_index] # Procedo al bilanciamento delle classi

print(f'Numero di osservazioni totali: {X_uncorr_balanced.shape}') # Ora la distribuzione delle classi è uniforme (214 per classe)
```

## Preprocessing - Oversampling Minor Class
Eseguo un oversampling sulla classe sotto-rappresentata creando l'80% di nuove osservazioni tramite SMOTE.

```python
from imblearn.over_sampling import SMOTE

X_test_0_ind = X_test.index  # Indice delle osservazioni del test set
available_X_0_indexes = [i for i in X_uncorr_norm[X_uncorr_norm['Bancarotta'] == 0].index if i not in list(X_test_0_ind)]  # Indice delle osservazioni per target = 0 senza quelle del test set

# Definisco il fattore di incremento per SMOTE
increasing_factor_smote = 1.8
dim_class_feature = int(X_train[y_train == 1].shape[0] * increasing_factor_smote)  # Convert to integer

# Select indices for class 0 and oversample class 1
X_train_1_ind = X_train[y_train == 1].index
X_train_0_ind = X_uncorr_norm.loc[available_X_0_indexes].sample(n=dim_class_feature, random_state=42).index

# Combine the indices
pre_smote_indexes = X_train_0_ind.union(X_train_1_ind)

x_cols = [i for i in X_uncorr_norm if i not in y_train.name]

# Perform SMOTE oversampling
oversample = SMOTE(random_state=42)
X_train_smote, y_train_smote = oversample.fit_resample(X_uncorr_norm.loc[pre_smote_indexes, x_cols], X_uncorr_norm['Bancarotta'].loc[pre_smote_indexes])

# Display results
print(X_train_smote.shape)
print(y_train_smote.value_counts())
```

## Feature Selection - SKB Evaluation Loop
Creo un loop per valutare, tramite l'approccio SKB (Select K Best), le features del dataset e salvo in le accuracies in una lista.

```python
from tqdm import tqdm
from sklearn.feature_selection import SelectKBest, RFE
from sklearn.feature_selection import f_classif
from sklearn.model_selection import KFold, cross_val_score 
from sklearn.linear_model import LogisticRegression

# SELECT K-BEST
def skb_selection(X, y, k):
    skb = SelectKBest(score_func=f_classif, k = k)
    skb = skb.fit(X, y) # Con fit trovo la relazione tra X e Y
    selected_features = X.columns[skb.get_support(indices=True)] # Restituisce l'indice delle features selezionate da SKB
    return X[selected_features] # Ritorno delle X originali solo quelle che SKB ritiene migliori

def skb_cross_validation_accuracy(X, y, k_fold = 5):
    # Creo il K-Fold cross-validator
    cv = KFold(n_splits=k_fold, shuffle=True, random_state=42)
    log_reg = LogisticRegression()

    # Eseguo la cross-validation misurando l'Accuracy
    scores = cross_val_score(
        log_reg, X, y, cv=cv, scoring='accuracy', n_jobs=-1
        )

    # Calcolo la media dell'Accuracy
    return scores.mean()
```

```python
# Calcolo l'accuracy per ogni k features
skb_accuracies = []
skb_new_col = []

for k in tqdm(range(1, len(X_train_smote.columns)+1)):
    X_selected = skb_selection(X_train_smote, y_train_smote, k)
    accuracy = skb_cross_validation_accuracy(X_selected, y_train_smote, 10)
    skb_accuracies.append(f'{accuracy}')
    new_elements = [i for i in X_selected.columns if i not in skb_new_col]
    if new_elements:  # Controlla se ci sono nuovi elementi
        skb_new_col.append(new_elements[0])
```

```python
# Stampo l'accuracy per ogni k features
for i, val in enumerate(skb_accuracies):
    print(f'{skb_new_col[i]}: {val}')
```

## Feature Selection - RFE Evaluation Loop
Creo un loop per valutare, tramite l'approccio RFE (Recursive Feature Elimination), le features del dataset e salvo in le accuracies in una lista.

```python
# RFE
log_reg = LogisticRegression(max_iter = 1000) # Inizializzo il modello di regressione logistica
def rfe_selection(X, y, k):
    rfe = RFE(estimator=log_reg, n_features_to_select=k)
    rfe = rfe.fit(X,y)
    selected_features = X.columns[rfe.support_]
    return X[selected_features]

def rfe_cross_validation_accuracy(X, y, k_fold = 5):
    # Creo il K-Fold cross-validator
    cv = KFold(n_splits=k_fold, shuffle=True, random_state=42)

     # Eseguo la cross-validation misurando l'Accuracy
    scores = cross_val_score(
        log_reg, X, y, cv=cv, scoring='accuracy', n_jobs=-1
        )
    
    # Calcolo la media dell'Accuracy
    return scores.mean()
```

```python
# Calcolo l'accuracy per ogni k features
rfe_accuracies = []
rfe_new_col = []

for k in tqdm(range(1, len(X_uncorr_balanced.columns))):
    X_selected = rfe_selection(X_train_smote, y_train_smote, k)
    accuracy = rfe_cross_validation_accuracy(X_selected, y_train_smote, 10)
    rfe_accuracies.append(f'{accuracy}')
    new_elements = [i for i in X_selected.columns if i not in rfe_new_col]
    if new_elements:  # Controlla se ci sono nuovi elementi
        rfe_new_col.append(new_elements[0])
```

```python
# Stampo l'accuracy per ogni k features
for i, val in enumerate(rfe_accuracies):
    print(f'{rfe_new_col[i]}: {val}')
```
