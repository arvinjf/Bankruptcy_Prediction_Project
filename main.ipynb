```python
first_limit = 0.4 # Soglia per correlazione positiva
second_limit = -0.4 # Soglia per correlazione negativa

def df_correl(data):
    key_columns = list()  # Lista vuota delle colonne che considererò nel dataset
    for i in range(data.shape[1]):  # Il for viene reiterato sul numero di colonne del dataset
        correl = data.iloc[:, i].corr(data.iloc[:, i-1])  # Calcolo la correlazione tra la colonna considerata e la sua precedente
        if (correl > 0 and correl < first_limit):  # Correlazione per valori positivi
            key_columns.append(data.iloc[:, i].name)
        elif (correl < 0 and correl > second_limit):  # Correlazione per valori negativi
            key_columns.append(data.iloc[:, i].name)
    return data[key_columns]

df = df_correl(df)
df
```
