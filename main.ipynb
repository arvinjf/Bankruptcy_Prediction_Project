```python
# Variance Inflation Factor
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(df):
    """
    Calculate VIF
    """
    vif = pd.DataFrame() # Inizializzo un nuovo dataframe
    vif["Variable"] = df.columns # Nella colonna Variable inserisco il nome delle colonne di df
    vif["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])] # Nella colonna VIF calcolo il VIF corrispondente a ogni colonna
    vif = vif.sort_values(by="VIF", ascending=False).reset_index(drop=True) # Ordini il dataframe in ordine decrescente in base a VIF
    return vif

vif = calculate_vif(df) # Calcolo il VIF sul dataset della bancarotta

# Count features with high VIF
vif_threshold = 5 # Imposto un limite per il VIF
features_to_remove = vif[vif['VIF'] > vif_threshold]['Variable'] # Seleziono le colonne che hanno un VIF maggiore del limite e le salvo in una variabile

# Remove features with VIF > 5
X_uncorr = df.drop(columns = features_to_remove) # Rimuovo dal dataframe le colonne selezionate precedentemente e quelle rimanenti le salvo in X_uncorr

vif = calculate_vif(X_uncorr) # Calcolo il VIF sulle colonne rimanenti
print(vif)
print("Features removed:", features_to_remove.shape[0]) # Visualizzo quante colonne sono state rimosse
print("Remaining features:", X_uncorr.shape[1]) # Visualizzo quante colonne sono rimaste
 
```
