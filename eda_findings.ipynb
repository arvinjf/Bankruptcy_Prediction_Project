### EDA - Feature Engineering:
Le features che presentano una elevata dispersione dei dati e una scala di misura diversa dalle altre sono:
- Tasso di Spesa Operativa
- Aliquota delle spese di ricerca e sviluppo
- Aliquota degli interessi sul debito
- Tasso crescita valore netto
- Debito totale/ Patrimonio netto totale

Queste features sono state trasformate in due modi:
- Square Root Transformation per ridurre la dispersione dei dati
- Robust Scaler per uniformare la scala di misura, mantenendo la proporzione tra i dati e attenuando l'effetto distorsivo degli outliers non rimossi
